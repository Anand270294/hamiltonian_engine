```python
obj1 = 'x_1 + x_2 + x_3'
v1 = ['x_1', 'x_2', 'x_3']

p1 = phs_ham(obj1, v1)
p1.get_objFun()
```

```python
import numpy as np
import networkx as nx 
import matplotlib.pyplot as plt 
from   matplotlib import cm
from   matplotlib.ticker import LinearLocator, FormatStrFormatter

#Generate a graph here using networkx
# Number of Nodes
n     = 3
V     = np.arange(0,n,1)
E     =[(0,1,1.0),(0,2,1.0),(1,2,1.0)] 
G     = nx.Graph()
G.add_nodes_from(V)
G.add_weighted_edges_from(E)

fig = plt.figure()
ax1 = plt.subplot2grid((1, 1), (0, 0))
labels = nx.get_edge_attributes(G,'weight')

nx.draw(G,with_labels=True, pos=nx.circular_layout(G), node_color='y', edge_color='g')
nx.draw_networkx_edge_labels(G,nx.circular_layout(G),edge_labels=labels)
```

```python
gammas = [0.1, 0.2]
betas  = [0.1, 0.2]
p2.perEdgeMap(gammas, 2, G, False, True)
circuit2 = p2.get_quantumCircuit()
circuit2.draw(output='mpl')
```

```python
mx_ham = mix_ham()

p2_map = p2.qubit_map 

mx_ham.generalXMixer(betas, 2, p2_map, True)
circuit2_mix = mx_ham.get_quantumCircuit()
circuit2_mix.draw(output='mpl')
```

```python
full_cir = p2 / mx_ham
full_cir.draw(output='mpl')
```
